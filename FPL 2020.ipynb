```python
import pandas as pd
```

 Loading the dataset

```python
df=pd.read_csv("cleaned_players.csv")
df1=pd.read_csv("players_raw.csv")
df3=pd.read_csv("teams.csv")
```

```python
df.head()
```

```python
df.columns
```

Lets add a new column combining first and last name 

```python
df["Name"]=df["first_name"]  + " " + df["second_name"]
df.head()
```

```python
df1["Name"]=df1["first_name"]  + " " + df1["second_name"]
df1=df1[["Name","team"]]
df1.head()
```

Lets merge the above two dataframes

```python
df2=df.merge(df1,on="Name")
df2
```

```python
df3=df3[["id","team_name"]]
df3.head()
```

Lets combine all 3 dataframes and start analyzing

```python
fpl=df2.merge(df3,left_on="team",right_on="id")
fpl.head()
```

# Which player has best Goal per 90 min record?

```python
df["Points_Per_90"]=df["minutes"]/90
df["Points_Per_Match"]=df["total_points"]/df["Points_Per_90"]
avg1=df[df["minutes"] > 700]
ee=avg1.sort_values(by="Points_Per_Match",ascending=False).head(20)
```

Lets import the viz libraries

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.figure(figsize=(15,10))
plt.title("Players with highest points Per 90 mins")
sns.barplot(x=ee.Points_Per_Match, y=ee.Name);
```

# Goalkeepers with most CS

```python
gk=fpl.sort_values(by="total_points",ascending=False)
goal=gk[gk["element_type"]=='GK']
gk_top=goal[["Name","total_points","team_name"]].head(20)

plt.figure(figsize=(15,10))
sns.barplot(x=gk_top.total_points, y=gk_top.Name)
plt.title("Goalkeepers with most points")
```

```python
mid=fpl.sort_values(by="total_points",ascending=False)
mid=mid[["Name","total_points","team_name"]].head(20)

plt.figure(figsize=(12, 6))
plt.xticks(rotation=90)
sns.barplot(x=mid.Name, y=mid.total_points)
plt.title("Mid with most points")
```

# Player with most Points by each team

```python
fpl["rank"]=fpl.groupby("team_name")['total_points'].rank(ascending=False)
rnk=fpl[fpl["rank"]==1]
rnk[["Name","team_name","total_points"]]
```

```python
fwd=fpl.sort_values(by="total_points",ascending=False)
fd=fwd[fwd["element_type"]=='FWD']
fd=fd[["Name","total_points","team_name"]].head(20)
```

```python
plt.figure(figsize=(12, 6))
plt.xticks(rotation=75)
sns.barplot(x=fd.Name, y=fd.total_points)

plt.title("Forwards with most points")
```

```python
mid=fpl.sort_values(by="total_points",ascending=False)
md=mid[mid["element_type"]=='MID']
md=md[["Name","total_points","team_name"]].head(20)

plt.figure(figsize=(12, 6))
plt.xticks(rotation=75)
sns.barplot(x=md.Name, y=md.total_points)
plt.title("Mid with most points")
```

```python
deff=fpl.sort_values(by="total_points",ascending=False)
df=deff[deff["element_type"]=='DEF']
df=df[["Name","total_points","team_name"]].head(20)

plt.figure(figsize=(15,10))

sns.barplot(x=df.total_points, y=df.Name)
plt.xlabel('Points')
plt.title("Defenders with most points")
```
